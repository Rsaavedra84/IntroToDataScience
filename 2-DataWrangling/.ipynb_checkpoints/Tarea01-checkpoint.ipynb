### P. 2.1 (1.5 puntos)

Programe un método de analisis de componentes principales utilizando la decomposición en autovalores de la matriz de correlación

\begin{eqnarray}
\boldsymbol{M}_x & = & \frac{1}{N-1}\boldsymbol{X}\boldsymbol{X}^\top\\
& = & \boldsymbol{U}\boldsymbol{\Sigma}^2\boldsymbol{U}^\top
\end{eqnarray}

```python
class PCA_M:
    def __init__ (self, n_components):
        self.n_components = n_components
        
    def fit (self, data):
        mu = data.mean(axis=0)
        X = data - mu #centered data
        N = X.shape[0] #number of obervations
        Mx = np.dot(X, X.transpose())/(N-1.) #Covariance matrix DxN
        l, U = np.linalg.eig(Mx) #autovalores y autovectores
        l = l.real
        N_pos = (l > 1e-10).sum()

        if (N_pos < self.n_components):
            self.n_components = N_pos
            
        U = U[:, (l > 1e-10)].real #2-dimension
        l = l[l > 1e-10]
        Sigma_inv = np.diagflat(np.sqrt(1./l))
        V = np.dot(np.dot(X.transpose(), U),Sigma_inv)/np.sqrt(N-1)
        self.components_ = V.transpose()[:self.n_components]
        
    def transform (self, data):
        return np.dot(self.components_, data.transpose())
    def inverse_transform (self, proj):
        return np.dot(proj.transpose(), self.components_)
```

```python
# test code

covariance = [[1., 1.3], [1.3, 2.]]
#covariance = [[0.5, 0.7], [0.7, 1.]]
mean = [1, 2]
N = 1000
samples = np.random.multivariate_normal(mean, covariance, N)
# lets define a linear model
m = np.sqrt(covariance[1][1]/ covariance[0][0])
y0 = mean[1]-m*mean[0]
#print m, y0
x = np.linspace (samples[:, 0].min(), samples[:, 0].max(), 30)
y = x * m + y0

# ---------
# The following code should be modified with
pca = PCA_M (n_components=2)

pca.fit (samples)
comp = pca.transform(samples)
u = pca.components_
print u

pl.clf()
pl.figure(figsize = (5,5))
pl.plot (samples[:, 0], samples[:, 1], "+")
pl.plot (x, y)
for i in range(u.shape[0]):
    pl.arrow(mean[0], mean[1], u[i, 0]*3, u[i, 1]*3, head_width=0.5, head_length=0.5, fc='g', ec='g')
pl.xlim ([-3, 7])
pl.ylim ([-3, 7])
pl.show()

```

### P. 2.2

Para el ejemplo de las "Olivetti faces" compare su codigo versus NMF (utilizar NMF de scikit-learn). Para esto, descomponga y reconstruya con distinto número de componentes utilizando ambos métodos y calcule el error cuadrático medio de sus reconstrucciones:

\begin{equation}
MSE = \frac{1}{N}\sum_i (y_{\mathrm{rec}} - y_{\mathrm{real}})^2.
\end{equation}

Grafique su $MSE$ en función del número de componentes y concluya cuál reconstruye mejor para los rangos que Ud. estudió.

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import NMF
rng = np.random.seed(0)

# Load faces data
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.data

n_samples, n_features = faces.shape

# global centering
faces_centered = faces - faces.mean(axis=0)

n_row, n_col = 2, 3
n_components = n_row * n_col
ns = 2**np.arange(1, 8)
#PCA first

mse = []
for n in ns:
   # print "running for ", n
    pca = PCA_M(n_components=n)
    pca.fit(faces)
    rec_pca = pca.inverse_transform(pca.transform(faces))
    e = ((rec_pca - faces)**2).sum()/(64*64)
    mse.append(e)

#NMF
mse_nmf = []
for n in ns:
    #print "running for ", n
    nmf = NMF(n_components=n)
    nmf.fit(faces)
    rec_nmf = nmf.inverse_transform(nmf.transform(faces))
    e2 = ((rec_nmf - faces)**2).sum()/(64*64)
    mse_nmf.append(e2)

    
```

```python
labels = ['2','4','8','16','32','64','128']

pl.figure(figsize=(15, 5))
pl.xlabel("Numbers of components")
pl.ylabel("MSE")
pl.plot(labels, mse, 'k--', label='mse of pca')
pl.plot(labels, mse_nmf, label='mse of nmf')

legend = pl.legend(loc='upper right', shadow=False)
pl.show()
```

Faces dataset has only positive components (pixels). Then, in this case, nmf rebuild better than pca
