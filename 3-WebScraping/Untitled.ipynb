# WEB SCRAPING 

````python
import requests # to obtain html data
import bs4      # beautifulsoup4 to parse the html content
````

The first step is to get the content related to an specific web page. In this example we will use the computer science department page of University of Concepcion.

````python
res = requests.get('http://www.inf.udec.cl/postgrado-2/oferta-de-asignaturas/')
````

````python
res.text
````

You can see the entire html (as plain text) content associated with the request. In order to parser and exctract information from this document we shall use [BeautifulSoup](https://www.crummy.com/software/BeautifulSoup/bs4/doc/).<br>
Using the ```bs4.BeautifulSoup(text, struct_of_data)``` method we can initialize an object which allow us to request and work with all tools that beatigulsoup give us.

````python
soup = bs4.BeautifulSoup(res.text, 'lxml')
````

Now we can select everything (in the html tags) that we need. For example we can select **the title** section

````python
title = soup.select('title')
````

The select function returns a list of objects. Alternatively, you can delete tags using the ```.getText()``` by selecting the  corresponding item into the list. In this we only have an element (the page title)

````python
title[0].getText()
````
